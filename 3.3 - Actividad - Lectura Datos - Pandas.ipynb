**Ejercicio de map**

Calcular la longitud del contenido de todas las celdas del datafame usando **lambda** y la operación **map**

```
import pandas as pd

leng = lambda x: len(x)

df = pd.DataFrame({'Nombre': ['Luz', 'Eunice','Ignacio','José'], 'Apellido': ['Angeles', 'Ochoa','Treviño','Zapata']})
df.applymap(leng)
```
